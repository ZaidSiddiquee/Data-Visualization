```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

First Visualization

```python
arr1 = np.arange(1,10)
arr1
```

```python
arr2 = arr1**2
arr2
```

```python
plt.plot(arr1,arr2)
```

```python
plt.plot(arr2,arr1)
```

Axis Labels

```python
plt.plot(arr1,arr2)
plt.xlabel("arr1")
plt.xlabel("arr2")
plt.title("arr1 vs arr2")
plt.show()
```

Subplot

```python
plt.subplot(1,2,1)
plt.plot(arr1,arr2, "r--")
plt.subplot(1,2,2)
plt.plot(arr2,arr1,  "r--")
```

```python
plt.subplot(2,1,1)
plt.plot(arr1,arr2, "r--")
plt.subplot(2,1,2)
plt.plot(arr2,arr1,  "r--")
```

Object Oriented Way

```python
#making figure
fig = plt.figure()

#set axes
axes1 = fig.add_axes([0.2,0.2,0.9,0.9])  #left,bottom,width,height

#plotting
axes1.plot(arr1,arr2)
axes1.set_xlabel("arr1")
axes1.set_ylabel("arr2")
axes1.set_title("arr1 vs arr2")
plt.show()
```

```python
fig = plt.figure()


axes1 = fig.add_axes([0.1,0.1,0.8,0.8])  #left,bottom,width,height

axes1.plot(arr1,arr2)
axes1.set_xlabel("arr1")
axes1.set_ylabel("arr2")
axes1.set_title("arr1 vs arr2", loc = "center")
```

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
fig = plt.figure()


axes1 = fig.add_axes([0.1,0.1,0.8,0.8])  #left,bottom,width,height

axes1.plot(arr1,arr2)
axes1.set_xlabel("arr1")
axes1.set_ylabel("arr2", fontdict = font2)
axes1.set_title("arr1 vs arr2", loc = "center",fontdict = font1)
```

```python
fig,axes = plt.subplots()
axes.plot(arr1,arr2,"r.", label = "arr1 vs arr2")
axes.plot(arr2,arr2,"g--", label = "arr2 vs arr1")
axes.legend()
```

```python
ypoints= np.array([3,8,10,9,5,8])
plt.plot(ypoints, marker = "o")
plt.show()
```

```python
plt.plot(ypoints, marker = "*", ms = 20)
plt.show()
```

```python
plt.plot(ypoints, marker = "*", ms = 40,mew = 2, mec = "r", mfc = "hotpink")  #ms = markersize, mew = markeredgewidth,  mec = markeredgecolor, mfc = markerfacecolor
plt.show()
```

```python
fig,axes = plt.subplots()
axes.plot(arr1,arr2,"r.", label = "arr1 vs arr2", ms = 30, mew = 2, mec = "r", mfc = "hotpink")
axes.plot(arr2,arr2,"g--", label = "arr2 vs arr1", ms = 30, mew = 2, mec = "b", mfc = "r")
plt.show()
```

Scatter Plot

```python
x = np.array([5,10,15,20,25,30,35,40])
y= np.array([1,2,3,4,5,6,7,8])
plt.xlabel("sales")
plt.ylabel("Profit")
plt.scatter(x,y, color = "hotpink")
x1 = np.array([6,10,16,20,26,37,35,40])
y1 = np.array([1,2,3,4,5,6,7,8])
plt.scatter(x1,y1)

plt.show()
```

```python
x = np.array([5,10,15,20,25,30,35,40])
y= np.array([1,2,3,4,5,6,7,8])
plt.xlabel("sales")
plt.ylabel("Profit")
sizes = np.array([20,50,100,200,500,1000,60,90])
plt.scatter(x,y, color = "hotpink", s = sizes, alpha = 0.5)
x1 = np.array([6,10,16,20,26,37,35,40])
y1 = np.array([1,2,3,4,5,6,7,8])
plt.scatter(x1,y1)

plt.show()
```

Bar chart

```python
x = np.array(["Chennai","Mumbai","Delhi","Kolkata"])
y = np.array([3,4,5,6])
plt.bar(x,y, width = 0.5)
plt.show()
```

```python
x = np.array(["Chennai","Mumbai","Delhi","Kolkata"])
y = np.array([3,4,5,6])
plt.barh(x,y, color = "red", height = 0.5) # for horizontal
plt.show()
```
